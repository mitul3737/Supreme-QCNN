```python
import pennylane as qml
import numpy as np
from scipy.stats import entropy
from collections import Counter
from scipy.special import rel_entr
import matplotlib.pyplot as plt

def U_TTN(params, layers, wires):  # 2 params cc_1
    for l in range(layers):
        offset = l * 2
        qml.RY(params[0], wires=wires[0])
        qml.RY(params[1], wires=wires[1])
        qml.CNOT(wires=[wires[0], wires[1]])


def U_5(params, layers, wires):  # 10 params CC_7
    for l in range(layers):
        offset = l * 10
        qml.RX(params[0], wires=wires[0])
        qml.RX(params[1], wires=wires[1])
        qml.RZ(params[2], wires=wires[0])
        qml.RZ(params[3], wires=wires[1])
        qml.CRZ(params[4], wires=[wires[1], wires[0]])
        qml.CRZ(params[5], wires=[wires[0], wires[1]])
        qml.RX(params[6], wires=wires[0])
        qml.RX(params[7], wires=wires[1])
        qml.RZ(params[8], wires=wires[0])
        qml.RZ(params[9], wires=wires[1])


def U_6(params, layers, wires):  # 10 params CC_8
    for l in range(layers):
        offset = l * 10
        qml.RX(params[0], wires=wires[0])
        qml.RX(params[1], wires=wires[1])
        qml.RZ(params[2], wires=wires[0])
        qml.RZ(params[3], wires=wires[1])
        qml.CRX(params[4], wires=[wires[1], wires[0]])
        qml.CRX(params[5], wires=[wires[0], wires[1]])
        qml.RX(params[6], wires=wires[0])
        qml.RX(params[7], wires=wires[1])
        qml.RZ(params[8], wires=wires[0])
        qml.RZ(params[9], wires=wires[1])


def U_9(params, layers, wires):  # 2 params cc_2
    for l in range(layers):
        offset = l * 2
        qml.Hadamard(wires=wires[0])
        qml.Hadamard(wires=wires[1])
        qml.CZ(wires=[wires[0], wires[1]])
        qml.RX(params[0], wires=wires[0])
        qml.RX(params[1], wires=wires[1])


def U_13(params, layers, wires):  # 6 params cc_4
    for l in range(layers):
        offset = l * 6
        qml.RY(params[0], wires=wires[0])
        qml.RY(params[1], wires=wires[1])
        qml.CRZ(params[2], wires=[wires[1], wires[0]])
        qml.RY(params[3], wires=wires[0])
        qml.RY(params[4], wires=wires[1])
        qml.CRZ(params[5], wires=[wires[0], wires[1]])


def U_14(params, layers, wires):  # 6 params cc_5
    for l in range(layers):
        offset = l * 6
        qml.RY(params[0], wires=wires[0])
        qml.RY(params[1], wires=wires[1])
        qml.CRX(params[2], wires=[wires[1], wires[0]])
        qml.RY(params[3], wires=wires[0])
        qml.RY(params[4], wires=wires[1])
        qml.CRX(params[5], wires=[wires[0], wires[1]])


def U_15(params, layers, wires):  # 4 params cc_3
    for l in range(layers):
        offset = l * 4
        qml.RY(params[0], wires=wires[0])
        qml.RY(params[1], wires=wires[1])
        qml.CNOT(wires=[wires[1], wires[0]])
        qml.RY(params[2], wires=wires[0])
        qml.RY(params[3], wires=wires[1])
        qml.CNOT(wires=[wires[0], wires[1]])


def U_SO4(params, layers, wires):  # 6 params CC_6
    for l in range(layers):
        offset = l * 6
        qml.RY(params[0], wires=wires[0])
        qml.RY(params[1], wires=wires[1])
        qml.CNOT(wires=[wires[0], wires[1]])
        qml.RY(params[2], wires=wires[0])
        qml.RY(params[3], wires=wires[1])
        qml.CNOT(wires=[wires[0], wires[1]])
        qml.RY(params[4], wires=wires[0])
        qml.RY(params[5], wires=wires[1])


def U_SU4(params, layers, wires): # 15 params CC_9
    for l in range(layers):
        offset = l * 15
        qml.U3(params[0], params[1], params[2], wires=wires[0])
        qml.U3(params[3], params[4], params[5], wires=wires[1])
        qml.CNOT(wires=[wires[0], wires[1]])
        qml.RY(params[6], wires=wires[0])
        qml.RZ(params[7], wires=wires[1])
        qml.CNOT(wires=[wires[1], wires[0]])
        qml.RY(params[8], wires=wires[0])
        qml.CNOT(wires=[wires[0], wires[1]])
        qml.U3(params[9], params[10], params[11], wires=wires[0])
        qml.U3(params[12], params[13], params[14], wires=wires[1])




list_of_circuits = [
    U_TTN,
    U_5,
    U_6,
    U_9,
    U_13,
    U_14,
    U_15,
    U_SO4,
    U_SU4

]
```

```python
import pennylane as qml
import numpy as np
from scipy.special import rel_entr
import matplotlib.pyplot as plt

# Bin setup
bins_list = [i / 75 for i in range(76)]
bins_x = [(bins_list[i] + bins_list[i+1]) / 2 for i in range(75)]

# Haar distribution
def P_harr(l, u, N):
    return (1 - l) ** (N - 1) - (1 - u) ** (N - 1)

P_harr_hist = np.array([P_harr(bins_list[i], bins_list[i+1], 16) for i in range(75)])
P_harr_hist /= P_harr_hist.sum()  # Normalize

# Device setup (removed parallel=True)
dev = qml.device("default.qubit", wires=4, shots=1000)

# Parameter counts per circuit
params_per_layer_map = {
    U_TTN: 2, U_5: 10, U_6: 10, U_9: 2, U_13: 6,
    U_14: 6, U_15: 4, U_SO4: 6, U_SU4: 15
}

expressivity_values = []
nlayers_list = [1, 2, 3, 4, 5]
nparam = 2000

for circuit in list_of_circuits:
    circuit_expressivities = []
    params_per_layer = params_per_layer_map[circuit]
    
    for layers in nlayers_list:
        total_params = params_per_layer * layers
        
        @qml.qnode(dev)
        def qnode(params):
            circuit(params, layers=layers, wires=range(4))
            return qml.probs(wires=range(4))
        
        # Initialize all probabilities array
        overlap_probabilities = np.zeros(nparam)
        
        for i in range(nparam):
            params = np.random.uniform(0, 2*np.pi, size=total_params)
            probs = qnode(params)
            overlap_probabilities[i] = probs[0]  # Probability of |0000
        
        weights = np.ones_like(overlap_probabilities) / len(overlap_probabilities)
        P_hist, _ = np.histogram(overlap_probabilities, bins=bins_list, 
                                weights=weights, range=[0, 1])
        
        # Stable KL calculation
        epsilon = 1e-12
        kl_pq = rel_entr(np.clip(P_hist, epsilon, None), 
                       np.clip(P_harr_hist, epsilon, None))
        average_kl = np.sum(kl_pq)
        circuit_expressivities.append(average_kl)
        print(f"{circuit.__name__}, Layers {layers}, KL Divergence: {average_kl:.4f}")
    
    expressivity_values.append(circuit_expressivities)

# Enhanced plotting
plt.figure(figsize=(14, 8))
for i, (circuit, kl_values) in enumerate(zip(list_of_circuits, expressivity_values)):
    plt.plot(nlayers_list, kl_values, marker='o', linestyle='--', 
            label=f'{circuit.__name__}')

plt.xlabel('Number of Layers', fontsize=12)
plt.ylabel('KL Divergence to Haar', fontsize=12)
plt.title('Circuit Expressivity Comparison', fontsize=14)
plt.xticks(nlayers_list)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
